### <span style="color: #007bff;">Exploratory Data Analysis on the World's Billionaires Dataset</span>

As part of this project, we conduct an exploratory data analysis (EDA) on a dataset containing information about the world's billionaires. This dataset provides insights into the demographics, sources of wealth, and industries associated with billionaire individuals globally.

#### Dataset Overview:
The dataset includes information such as the rank of billionaires, their names, net worth, age, country/territory, primary source of wealth, and industry. With over 2600 entries, this dataset offers a comprehensive view of the billionaire landscape.

#### Objectives:
1. Explore the distribution of billionaires by age, country/territory, industry, and net worth.
2. Analyze the top sources of wealth among billionaires and identify trends within specific industries.
3. Visualize key insights to gain a better understanding of the billionaire population globally.

#### Methodology:
We'll utilize Python libraries such as Pandas, Matplotlib, and Seaborn for data manipulation, visualization, and analysis. Through descriptive statistics, visualizations, and interpretation, we aim to uncover patterns, trends, and outliers within the dataset.

Let's embark on this journey to explore the fascinating world of billionaires and uncover insights that shed light on wealth distribution, industry dynamics, and global economic trends.

# <span style="color: #007bff;"> 2. Checking out missing values or null values:</span>

</div>

```python
df.isnull().sum()
```

```python


# Plot histogram of ages
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Replace missing values with the mean age
mean_age = df['Age'].mean()
df['Age'].fillna(mean_age, inplace=True)
```

```python
df.isnull().sum()
```

<div style="background-color: #f9f2f4; padding: 10px; border-radius: 5px;">

# <span style="color: #007bff;"> 3. Exploratory Data Analysis:</span>

</div>
    

```python
# Remove currency symbol and 'B' from 'Net Worth' column and convert to numerical values
df['Net Worth'] = df['Net Worth'].replace({'\$': '', 'B': ''}, regex=True).astype(float)

# Sort the DataFrame by 'Net Worth' column in descending order
top_5_billionaires = df.sort_values(by='Net Worth', ascending=False).head(5)

# Display the top 5 billionaires DataFrame view
top_5_billionaires[['Rank', 'Name', 'Net Worth', 'Age', 'Country | Territory', 'Source', 'Industry']]
```

```python
# Create a boxplot of ages with customized colors
plt.figure(figsize=(8, 6))
plt.boxplot(df['Age'].dropna(), vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue', color='black'), medianprops=dict(color='orange'))
plt.title('Age Distribution of Billionaires')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```

<div style="background-color: #f9f2f4; padding: 10px; border-radius: 5px;">

### <span style="color: #007bff;">Age Distribution of Billionaires</span>

- Most billionaires are between 55 and 70 years old, with a peak at 65-70 years old.
- There are relatively few billionaires under the age of 40, indicating that younger individuals are less represented among billionaires.
- The number of billionaires gradually declines as age increases, and there are very few billionaires over the age of 80.
- The age distribution of billionaires is skewed to the right, indicating that there are more billionaires in older age groups than in younger age groups.

</div>



```python
# Set the number of top countries/territories to include in the pie chart
top_n = 10

# Count the number of billionaires per country/territory
country_counts = df['Country | Territory'].value_counts().head(top_n)

# Plot the distribution of billionaires by country/territory using a pie chart
plt.figure(figsize=(7, 9))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=160, colors=plt.cm.tab10.colors)
plt.title('Distribution of Billionaires by Country/Territory (Top {})'.format(top_n))
plt.axis('equal')
plt.show()
```

<div style="background-color: #f9f2f4; padding: 10px; border-radius: 5px;">

### <span style="color: #007bff;">Dominant Presence of US Billionaires:</span>

- The United States dominates the chart with a staggering 38.1% share of billionaires, which amounts to about 691 individuals. This signifies a significant lead over all other countries.
- China, previously placed first in my misinterpretation, takes the second position with 25.5%, translating to an estimated 469 billionaires. While substantial, it falls considerably behind the US figure.
- India maintains its third position with 8.8%, representing approximately 187 billionaires.
- Germany remains in fourth place with 6.5%, equivalent to around 144 billionaires.
- The remaining six countries (Hong Kong, Russia, Italy, Canada, UK, Taiwan) collectively hold around 21.1% of the total, with individual shares ranging from 2.7% to 5.4%, roughly translating to 60-120 billionaires each.

### <span style="color: #007bff;">Key Observations:</span>

- The pie chart highlights the remarkable concentration of wealth in the United States, holding almost twice the number of billionaires compared to China, the runner-up.
- The top three countries (US, China, India) contribute nearly 73% of the total billionaires represented, indicating a significant disparity in global wealth distribution.
- While developed countries generally hold more billionaires, the presence of several non-Western economies (India, Hong Kong) showcases a diversifying landscape.

</div>


```python
# Count the number of billionaires per industry
industry_counts = df['Industry'].value_counts()

# Plot the distribution of billionaires by industry
plt.figure(figsize=(12, 6))
industry_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Billionaires by Industry')
plt.xlabel('Industry')
plt.ylabel('Number of Billionaires')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

<div style="background-color: #f9f2f4; padding: 10px; border-radius: 5px;">

### <span style="color: #007bff;">Key Takeaways from Industry Distribution:</span>

- The top three industries for billionaires are finance & investments, manufacturing, and technology. These industries together account for nearly 40% of the total number of billionaires on the chart.
- Finance & investments is the clear leader, with 372 billionaires, which is about 14% of the total. This includes individuals who made their fortunes in areas like banking, hedge funds, and private equity.
- Manufacturing comes in second with 324 billionaires, or about 12% of the total. This reflects the continued importance of traditional industries like automobiles, chemicals, and consumer goods.
- Technology is close behind with 313 billionaires, also around 12% of the total. This encompasses a wide range of tech subsectors, including software, internet, and electronics.
- Other notable industries on the chart include fashion & retail, food & beverage, healthcare, and real estate. Each of these industries has over 100 billionaires.

</div>


```python
# Plot a boxplot of net worth
plt.figure(figsize=(8, 6))
plt.boxplot(df['Net Worth'].dropna(), vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue', color='black'), medianprops=dict(color='orange'))
plt.title('Net Worth Distribution of Billionaires')
plt.xlabel('Net Worth (in billion USD)')
plt.grid(True)
plt.show()
```

<div style="background-color: #f9f2f4; padding: 10px; border-radius: 5px;">

### <span style="color: #007bff;">General Observations from the Box Plot:</span>

- The box plot displays the net worth distribution of the world's billionaires. This means it shows how the net worth values are spread out.
- The horizontal axis likely represents net worth in billions of dollars.
- The box in the center represents the middle 50% of the data. The line in the middle of the box is likely the median net worth, which means half of the billionaires have a net worth above this value and half have a net worth below it.
- The whiskers extending from the box represent the remaining 25% of the data on either side. The ends of the whiskers typically represent the minimum and maximum net worth values, but they can also be defined differently depending on the specific way the box plot is constructed.

</div>


```python
# Create a scatter plot to visualize the relationship between age and net worth
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Net Worth', data=df, hue='Industry', palette='Set1', alpha=0.7)
plt.title('Relationship Between Age and Net Worth by Industry')
plt.xlabel('Age')
plt.ylabel('Net Worth (in billion USD)')
plt.grid(True)
plt.legend(title='Industry', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

<div style="background-color: #f9f2f4; padding: 10px; border-radius: 5px;">

### <span style="color: #007bff;">Overall Trends:</span>

- There's a positive correlation between age and net worth across most industries. As billionaires age, their net worth generally increases. This might be due to factors like accumulating wealth over time, holding leadership positions in companies, or making strategic investments.
- The technology industry has the youngest billionaires on average, with a cluster of data points concentrated in the lower left portion of the chart. This suggests that many individuals in this sector achieved billionaire status at a young age, possibly through successful tech startups or ventures.
- Several industries, including finance & investments, manufacturing, and fashion & retail, exhibit a wider range of net worth across different age groups. This indicates that some individuals in these fields amassed their wealth early, while others reached billionaire status later in life.

### <span style="color: #007bff;">Industry-Specific Observations:</span>

- **Finance & investments:** This industry boasts some of the highest net worth individuals, even among younger age groups. This aligns with the stereotype of successful financiers accumulating significant wealth early in their careers.
- **Manufacturing:** This sector shows a steady increase in net worth with age, suggesting that building wealth in manufacturing often takes time and experience.
- **Technology:** While the industry has younger billionaires, their net worth varies considerably. This reflects the diverse paths to wealth creation in the tech sector, where some individuals achieve rapid success, while others build their fortunes over extended periods.
- **Fashion & retail:** This industry showcases a mix of billionaires who reached their peak net worth at different ages, indicating diverse career trajectories and wealth accumulation patterns.

### <span style="color: #007bff;">Important Caveats:</span>

- The chart only depicts a snapshot of the data for 2023. Net worth can fluctuate over time, and individual circumstances can influence wealth accumulation.
- The scatter plot doesn't account for factors beyond age and industry that might influence net worth, such as education, inheritance, or personal financial decisions.
- The data might not be perfectly representative of all the world's billionaires, as the source and methodology for selecting individuals might influence the results.

</div>


```python
# Get the top 10 sources of wealth
top_sources = df['Source'].value_counts().head(10)

# Plot the top sources of wealth
plt.figure(figsize=(10, 6))
top_sources.plot(kind='bar', color='darkblue')
plt.title('Top Sources of Wealth Among Billionaires')
plt.xlabel('Source of Wealth')
plt.ylabel('Number of Billionaires')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

<div style="background-color: #f9f2f4; padding: 10px; border-radius: 5px;">

### <span style="color: #007bff;">Interpretation:</span>

- **Real estate** is the most common source of wealth among billionaires, with a significant gap between it and the other sources. This could be due to various factors, such as the long-term value appreciation of real estate, rental income generation, and potential for development.
- **Investments** is the second most common source of wealth, followed by **diversified**, which likely refers to individuals with wealth across multiple industries or asset classes.
- Other notable sources of wealth include **pharmaceuticals**, **software**, **hedge funds**, **private equity**, **chemicals**, **retail**, and **manufacturing**. These diverse industries highlight the various paths individuals can take to accumulate significant wealth.

</div>


```python
# Get the top 10 billionaires
top_10_billionaires = df.head(10)

# Group by source of wealth and count the number of occurrences
source_counts = top_10_billionaires.groupby('Source').size().reset_index(name='Count')

# Plot the top sources of wealth for the top 10 billionaires
plt.figure(figsize=(12, 10))
plt.bar(source_counts['Source'], source_counts['Count'], color='skyblue')

# Annotate each bar with the name of the billionaire
for i, (source, count) in enumerate(zip(source_counts['Source'], source_counts['Count'])):
    plt.text(i, count, '\n'.join(top_10_billionaires[top_10_billionaires['Source'] == source]['Name']), ha='center', va='bottom', rotation=10)

plt.title("Top Sources of Wealth Among the World's Top 10 Billionaires")
plt.xlabel('Source of Wealth')
plt.ylabel('Number of Billionaires')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

<div style="background-color: #f9f2f4; padding: 10px; border-radius: 5px;">

### <span style="color: #007bff;">Interpretation:</span>

- **Dominant Tech and Retail Sectors:**  
The chart highlights the significant presence of individuals from the technology and retail sectors among the world's top 10 billionaires. Elon Musk (Tesla, SpaceX), Bernard Arnault (LVMH - luxury goods), Jeff Bezos (Amazon), Larry Ellison (Oracle), Steve Ballmer (Microsoft), Larry Page (Google), and Mark Zuckerberg (Meta) all built their fortunes in these industries. This reflects the immense wealth-generating potential of these sectors in the current global economy.

- **Diversification Beyond Tech and Retail:**  
While tech and retail are prominent, other sectors are also represented. Bill Gates (Microsoft) and Warren Buffett (Berkshire Hathaway) come from the software and investment industries, respectively. Mukesh Ambani's wealth stems from his diversified conglomerate Reliance Industries, and Carlos Slim Helu's fortune is primarily in telecom through América Móvil. This indicates that significant wealth creation can occur in various sectors beyond just tech and retail.

- **Individual Success Stories:**  
Each billionaire on the chart has a unique story of how they achieved their immense wealth. Elon Musk's ventures in electric vehicles and space exploration, Bernard Arnault's leadership in the luxury goods industry, and Jeff Bezos's e-commerce giant Amazon are just a few examples. These individuals' remarkable achievements and business acumen have propelled them to the top of the wealth pyramid.

- **Important Considerations:**  
It's important to remember that this chart only represents the top 10 richest people globally. Wealth distribution remains highly unequal, with a vast majority of individuals holding significantly less wealth than those depicted here.  
The sources of wealth listed might not capture the full picture, as some individuals might have diversified their holdings or have wealth from other sources not mentioned.  
This data represents a snapshot in time, and wealth rankings can fluctuate due to various factors like stock market performance and business ventures.
